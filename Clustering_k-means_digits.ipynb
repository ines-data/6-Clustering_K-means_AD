# Qu'est ce que le clustering ?

Le clustering est une méthode d’apprentissage automatique qui consiste à regrouper des points de données par similarité ou par distance. C’est une méthode d’apprentissage non supervisée.  
La but des algorithmes de clustering est de donner un sens aux données et d’extraire de la valeur à partir de grandes quantités de données structurées et non structurées. Ces algorithmes vous permettent de séparer les données en fonction de leurs propriétés ou fonctionnalités et de les regrouper dans différents clusters en fonction de leurs similitudes.

# quelle est la compléxité d'un probléme de clustering ?

Le clustering K-means est l’un des algorithmes de machine learning non supervisés les plus facile à comprendre et à utiliser. Aujourd’hui cet algorithme est l’un des plus répandus.
D’habitude, les algorithmes non supervisés font des inférences à partir de jeux de données en se servant uniquement des vecteurs d’entrée sans faire référence à des résultats connus ou étiquetés.  

Les K-means est une technique de classification par apprentissage automatique utilisée pour simplifier des ensembles de données volumineux en ensembles de données simples et plus petits. Des modèles distincts sont évalués et des ensembles de données similaires sont regroupés.   
La variable K représente le nombre de groupes dans les données.

# La difficile évaluation des clusters

En effet, le clustering est une tâche non supervisée. Il n’existe pas de données annotées contenant les clusters à trouver. Nous n’avons donc pas de clusters de référence
auxquels comparer les clusters obtenus. Des mesures de qualité des clusters existent et
peuvent être classées en deux catégories :  

– les mesures dites **internes**, qui évaluent en général la capacité d’un algorithme à
produire des clusters qui maximisent la similarité intra-cluster et minimisent la
similarité inter-cluster, elles ne nécessitent pas de données annotées (sans vérité terrain).  

– les mesures dites **externes**, qui comparent les résultats obtenus par l’algorithme
à des résultats attendus. Ces mesures nécessitent alors un ensemble de données
annotées donc dans une vérité terrain.

# Quelle sont les métrique utilisé pour le clustering ?

## Les mesures Internes

**Indice de Dunn :** est le rapport entre la distance maximale qui sépare deux éléments classés ensemble et la distance minimale qui sépare deux
éléments classés séparément.  
Un indice de Dunn élevé indique une distance inter-classes élevée et une distance intra-classes faible,
ce qui correspond à la définition d’une bonne partition des données.  
`from jqmcvi import base `   
`clus0 = prediction.loc[prediction.Species == 0]   
 clus1 = prediction.loc[prediction.Species == 1]   
 clus2 = prediction.loc[prediction.Species == 2]   
 cluster_list = [clus0.values, clus1.values, clus2.values] `
`print(base.dunn(cluster_list))`

**Silhouette :** est un indice pouvant être calculé pour chaque
objet, cluster ou pour un clustering entier. Cette mesure est comprise entre -1 et 1,
une valeur positive indique des clusters compacts et bien séparés.   
Pour un objet, la
silhouette permet d’évaluer si les objets qui appartiennent au même cluster qu’un objet
x sont plus proches de x que les objets des autres clusters.  
`sklearn.metrics.silhouette_score(X, labels, *, metric='euclidean', sample_size=None, random_state=None, **kwds)`

**Indice de Davies-Bouldin :** est un indice qui
favorise les clusters compacts et bien séparés dans l’espace des données. Une valeur
faible indique un clustering de bonne qualité.  
`sklearn.metrics.davies_bouldin_score( X , étiquettes )`

Le
problème des mesures internes est qu’elles ne valident pas que les clusters obtenus ont
un sens pour l’utilisateur qui visualisera les résultats. Cela peut être validé avec les
mesures externes.

## Les mesures Externes

**Indice de pureté :** L’indice de pureté évalue si les clusters contiennent des objets
d’une seule classe. L’indice est compris entre 0 et 1, la valeur 1 indiquant que les clusters
sont tous purs. L’inconvénient de l’indice de pureté est qu’il surévalue la qualité d’un
clustering avec un nombre important de classes. En effet, l’indice de pureté est maximal
si le nombre de clusters correspond au nombre d’individus.  

**Rand index :** Le Rand index est une mesure comparant deux clustering. Cet indice
correspond au ratio des paires de points pour lesquels les deux algorithmes donnent les
mêmes résultats. Plus le Rand index est élevé, plus les clusterings sont proches. Le Rand
index pénalise à la fois les faux positifs et les faux négatifs. L’inconvénient majeur du
Rand index est qu’il est difficilement comparable.  
`metrics.rand_score(labels_true, labels_pred)`

**Rand index ajusté :**  Une version corrigée du Rand index, le Rand
index ajusté, dont la valeur est comprise entre 0 et 1. C’est l’une de mesure d’évaluation
externe les plus couramment utilisées.  
`metrics.adjusted_rand_score(labels_true, labels_pred)`

**F-mesure :** La F-mesure évalue la propension d’un cluster à contenir à la fois tous
les objets d’une classe et seulement les objets de cette classe. Elle permet d’affecter
différents poids pour les faux positifs et les faux négatifs. De plus, elle prend en compte
le nombre de clusters générés, comparé au nombre réels de classe dans la vérité terrain.

```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

# Clustering k-means

```python
X = digits.images
#X = digits.data
print(X.shape)
X = digits.images.reshape(1797,64)
print(X.shape)
Y = digits.target
print(X.shape)
X
```

```python
model_0 = KMeans( init="k-means++", n_clusters=10, n_init=10, max_iter=300)
model_0.fit(X)
```

```python
model_0.inertia_
```

Ici on connait le nombre de cluster qui est 10, par contre si on ne le connaisait pas on peut le trouver soit avec la **méthode du coude**, ou le **coefficient de silhouette**

# Méthode du coude

```python
tab = []
for k in range(1, 30):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    tab.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 30), tab)
plt.xticks(range(1, 30))
plt.xlabel("Number of Clusters")
plt.ylabel("Tab")
plt.show()
```

```python
#La détermination du point de coude dans la courbe
point_coude = KneeLocator(range(1, 30), tab, curve="convex", direction="decreasing")

point_coude.elbow
```

ici on obtient 9 alors que dans notre data on sait que le nombre de clusters est 10

## Coefficient de silouhette

```python
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 20), silhouette_coefficients)
plt.xticks(range(2, 20))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

ici on obtient 9 alors que dans notre data on sait que le nombre de clusters est 10

```python
model = KMeans(n_clusters=10)
model.fit(X)
y_pred= model.predict(X)
```

```python
from sklearn import metrics
print("inertia = ", model.inertia_)
print("ARI = ",metrics.adjusted_rand_score(Y, y_pred))
print("NMI = ",metrics.normalized_mutual_info_score(Y,y_pred))
print("AMI = ",metrics.adjusted_mutual_info_score(Y,y_pred))
print("Accuracy = ",metrics.accuracy_score(Y, y_pred))
print("silouhette score = ",metrics.silhouette_score(X,y_pred,metric="euclidean"))
```

# K-means avec PCA = 2

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
model_3 = KMeans(init="k-means++", n_clusters=10, n_init=4)
y3_pred = model_3.fit_predict(X_pca)
```

```python
from sklearn import metrics

print("inertia = ", model_3.inertia_)
print("ARI = ",metrics.adjusted_rand_score(Y, y3_pred))
print("NMI = ",metrics.normalized_mutual_info_score(Y,y3_pred))
print("AMI = ",metrics.adjusted_mutual_info_score(Y,y3_pred))
print("Accuracy = ",metrics.accuracy_score(Y, y3_pred))
print("silouhette score = ",metrics.silhouette_score(X,y3_pred,metric="euclidean"))
```
